```python
#normal

arr = [1, 2, 3, 4, 5]

for i in range(len(arr)):
    if arr[i]%2==0:
        print(arr[i])
```

# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#simple example

x = np.arange(0,10)
y = np.arange(11, 21)

x, y
```

```python
plt.scatter(x,y) 
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter Plot")
```

```python
y = x**3
y
```

```python
##plt plot

plt.plot(x,y, 'r*', linestyle='dashed', linewidth=2, markersize=12)  #r* = red color
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2d diagram')
```

```python
#Creating subplots

plt.subplot(2,2,1)           #(nrows,ncolumns,index)
plt.plot(x, y, 'r--')
plt.subplot(2,2,2)
plt.plot(x, y, 'g*--')
plt.subplot(2,2,3)
plt.plot(x, y, 'bo')
plt.subplot(2,2,4)
plt.plot(x, y, 'go') 
```

```python
#Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,17]

plt.bar(x,y)
plt.bar(x2,y2, color = 'g')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
#Histogram

a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

```python
#Pie Chart

#Data

labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

#Plot
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%')

plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```

```python
#Box Plot

data = [np.random.normal(0,std,100) for std in range(1,4)]

#recangular box plot

plt.boxplot(data, vert=True, patch_artist=False)
```

# Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.shape
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
#Pair Plot

sns.pairplot(df)
```

```python
sns.pairplot(df, hue="time")
```

```python
#checking distribution of the features

sns.distplot(df['tip'])
```

```python
#JoinPlot -> relationship of numerical features

sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.countplot(x='sex', data=df)
```

```python
sns.countplot(y='sex', data=df)
```

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x="day",y="total_bill", hue="smoker", data=df)
```

```python
sns.violinplot(x="day", y="total_bill", data=df)
```
